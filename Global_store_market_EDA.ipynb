Converting  conlumn orderdate and ship date into date time object

```python
order_date = df['Order Date'].tolist()
```

```python
ship_date = df['Ship Date'].tolist()
```

```python
order_date_time = pd.to_datetime(order_date)
```

```python
ship_date_time = pd.to_datetime(ship_date)
```

```python
#now converting the date time object into series 
Order_date = pd.Series(order_date_time,name='Order_Date')
Ship_date = pd.Series(ship_date_time,name='Ship_Date')
```

```python
#drop unneccessary column
```

```python
df.drop(columns=['Row ID','Order Date','Ship Date','Customer ID','Postal Code'],inplace=True)
```

```python
df.head()
```

```python
df2 = pd.concat([df, Order_date, Ship_date], axis=1)
```

```python
df2.head()
```

Convert into integer

```python
df2['Shipping Cost'] = df2['Shipping Cost'].astype(int)
```

```python
df2['Discount'] = df2['Discount'].astype(int)
```

convert string into integer

```python
df2['Sales'] = df2['Sales'].str.replace('$','',regex=True)
df2.head()
```

```python
df2['Sales'] = df2['Sales'].str.replace(',','',regex=True)
df2.head()
```

```python
df2['Profit'] = df2['Profit'].str.replace('$','',regex=True)
```

```python
df2['Profit'] = df2['Profit'].str.replace(',','',regex=True)
```

```python
df2.head()
```

```python
df2['Profit'] = df2['Profit'].str.replace('(','',regex=True)
```

```python
df2['Profit'] = df2['Profit'].str.replace(')','',regex=True)
```

```python
df2['Sales'] = pd.to_numeric(df2['Sales'],downcast='signed')
```

```python
df2['Profit'] = pd.to_numeric(df2['Profit'],downcast='signed')
```

```python
df2.head()
```

```python
df2['Sales'] = df2['Sales'].astype(int)
```

```python
df2['Profit'] = df2['Profit'].astype(int)
```

#### 8. Find the overall rank of "India" where the 'Profit' is maximum under the category 'Technology'.
